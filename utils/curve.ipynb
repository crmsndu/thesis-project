```julia
using Plots
```

```julia
err_st_1 = [0.145, 0.115, 0.0813, 0.0554, 0.012, 6.65e-4, 1.07e-5, 4e-7, 1.25e-8]
x_st_1 = [1.2, 2.4, 3.6, 4.2, 4.8, 5.1, 5.4, 5.7, 6]
```

```julia
err_st_2 = [0.146, 0.118, 0.0866, 0.0644, 0.012, 3.12e-4, 8.81e-6, 2.82e-7, 1.18e-8]
x_st_2 = [1.2, 2.4, 3.6, 4.2, 4.8, 5.1, 5.4, 5.7, 6]
```

```julia
plot(x_st_1, err_st_1, xlabel="SNR (dB)", ylabel="BER", minorgrid=:true,
    yaxis=:log, label="Iteration(1)", legend=:bottomleft)
plot!(x_st_2, err_st_2, label="Iteration(3)")
```
